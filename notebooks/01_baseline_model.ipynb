# Baseline cultivated meat process model

This notebook runs the baseline scenario defined in `config/baseline.yaml`.

```python
from pathlib import Path

from cm_process_model import Flowsheet, load_config

config_path = Path("../config/baseline.yaml")
config = load_config(config_path)
flowsheet = Flowsheet(config)
results = flowsheet.simulate()
```

```python
for name, stream in results.items():
    print(f"{name:12s}  mass={stream.mass_kg:,.1f} kg  volume={stream.volume_L:,.0f} L")

print(f"\nDaily packaged product: {flowsheet.daily_product_kg:,.1f} kg")
```
